```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, f_classif, chi2
```

```python
data.head()
```

```python
df=pd.DataFrame(data=data)
df.head()
```

```python
X=data[['Distance','MaxSpeed','AvgSpeed','AvgMovingSpeed','FuelEconomy','TotalTime','MovingTime']]
Y=data['GoingTo']
```

```python
#F-test
from sklearn.feature_selection import SelectKBest 
from sklearn.feature_selection import f_classif 
fs = SelectKBest(score_func=f_classif, k=3) 
fit = fs.fit(X,Y) 
dfscores = pd.DataFrame(fit.scores_)  
dfcolumns = pd.DataFrame(df.columns)  

featureScores = pd.concat([dfcolumns,dfscores],axis=1) 
featureScores.columns = ['Selected_columns','Score_ANOVA']
print(featureScores.nlargest(3,'Score_ANOVA'))
```

```python
#Chi-square
sel_chi=SelectKBest(chi2,k=4)
X_train_chi=sel_chi.fit_transform(X,Y)
print(X_train_chi.shape)
print(sel_chi.get_support())
```

```python
from sklearn.feature_selection import RFE 
from sklearn.linear_model import LogisticRegression  
NUM_FEATURES = 3 
model = LogisticRegression(random_state=123) 
rfe = RFE(model, NUM_FEATURES) 
fit = rfe.fit(X,Y) 
print("Num Features: %s" % (fit.n_features_)) 
print("Selected Features: %s" % (fit.support_)) 
print("Feature Ranking: %s" % (fit.ranking_))
```
